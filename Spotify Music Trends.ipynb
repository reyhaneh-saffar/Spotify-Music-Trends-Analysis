# **Shapiro-Wilk Test**
Tests whether a data sample has a Gaussian distribution.

Assumption

Observations in each sample are independent and distributed identically.

Hypothesis

H0: the sample has a Gaussian distribution.

H1: the sample does not have a Gaussian distribution.

Here i'm gonna apply Shapiro-Wilk Test on some features

```python
from scipy.stats import shapiro
Danceability_df = df.dropna(subset=["Danceability"])
stat, p = shapiro(Danceability_df['Danceability'])
print('stat={0:.3f}, p={0:.3f}' .format(stat, p))
if p > 0.05:
    print('Probably Gaussian')
else:
    print('Probably not Gaussian')
```

# **Pearson’s Correlation Coefficient**
Tests whether a data sample is linearly separable.
Assumption

a) Observations in each sample are independent and distributed identically.

b) Observations are normally distributed.

c) Similar variance between independent variables

Hypothesis

H0: the samples are correlated.

H1: the sample does not have any correlation.

Here i'm gonna use this set on some features.

```python
from scipy.stats import pearsonr
sub1_df = copy_df.dropna(subset=["Danceability"])
sub2_df = copy_df.dropna(subset=["Energy"])
stat, p = pearsonr(sub1_df['Danceability'], sub2_df['Energy'])
print('stat={0:.3f}, p={0:.3f}'.format(stat, p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent')
```

# **ANOVA test**

**Do the top 10,000 songs on Spotify from 1960 to now have a statistically significant difference in popularity between decades?**
Using ANOVA test to determine whether there is a significant difference in popularity between the decades.


```python
copy_df['Album Release Date'].unique()
```

```python
copy_df['Popularity'].unique()
```

```python
import pandas as pd
from scipy.stats import f_oneway

# Extract the decade from the Album Release Date column
copy_df["decade"] = copy_df["Album Release Date"].apply(lambda x: str(x)[:3] + "0s")

# Perform the ANOVA test
f_value, p_value = f_oneway(copy_df[copy_df["decade"] == "1960s"]["Popularity"],
                             copy_df[copy_df["decade"] == "1970s"]["Popularity"],
                             copy_df[copy_df["decade"] == "1980s"]["Popularity"],
                             copy_df[copy_df["decade"] == "1990s"]["Popularity"],
                             copy_df[copy_df["decade"] == "2000s"]["Popularity"],
                             copy_df[copy_df["decade"] == "2010s"]["Popularity"])
print("F-value:", f_value)
print("P-value:", p_value)
```

The F-value is 36.44 and the p-value is 4.24e-37. Since the p-value is less than 0.05, we can reject the null hypothesis and conclude that there is a statistically significant difference in popularity between the decades. The F-value indicates the ratio of the variance between the groups to the variance within the groups. A higher F-value indicates a greater difference between the groups. In this case, the F-value is relatively high, indicating that there is a significant difference in popularity between the decades.

#  **Pearson correlation test**

**Is there a correlation between the danceability and energy of the top 10,000 songs on Spotify from 1960 to now?**
We will use a Pearson correlation test to determine whether there is a significant correlation between danceability and energy.

**Hypothesis**: The null hypothesis is that there is no significant correlation between danceability and energy. The alternative hypothesis is that there is a significant correlation between danceability and energy.

```python
new_df = copy_df.copy(deep=True)
```

```python
new_df['Danceability'].unique()
```

```python
new_df['Energy'].unique()
```

```python
import pandas as pd
from scipy.stats import pearsonr

# Clean the data
new_df = new_df.dropna(subset=["Danceability", "Energy"])

# Perform the Pearson correlation test
corr, p_value = pearsonr(new_df["Danceability"], new_df["Energy"])

print("Correlation coefficient:", corr)
print("P-value:", p_value)
```

The correlation coefficient is 0.135 and the p-value is 1.29e-41. Since the p-value is less than 0.05, we can reject the null hypothesis and conclude that there is a statistically significant correlation between danceability and energy. The correlation coefficient indicates the strength and direction of the linear relationship between two variables. A value of 0.135 indicates a weak positive correlation between danceability and energy.

# **T-test**

**Is there a statistically significant difference in the popularity of songs with different genres?**
 We will use a t-test to compare the popularity of songs in different genres.

**Hypothesis**: The null hypothesis is that there is no significant difference in popularity between the genres. The alternative hypothesis is that there is a significant difference in popularity between the genres.

```python
new_df['Artist Genres'].unique()
```

There are 2815 different genres represented in the dataset. I want to compare the popularity of songs in these five genres:
pop, rock, hip hop, country, jazz.

```python
import pandas as pd
from scipy.stats import ttest_ind

# Clean the data
new_df = new_df.dropna(subset=["Artist Genres", "Popularity"])

# Perform the t-test
popularity_by_genre = [new_df[new_df["Artist Genres"].str.contains(genre)]["Popularity"] for genre in ["pop", "rock", "hip hop", "country", "jazz"]]
t_value, p_value = ttest_ind(popularity_by_genre[0], popularity_by_genre[1])

# Print the results
print("T-value:", t_value)
print("P-value:", p_value)
```

The t-value is 3.37 and the p-value is 0.00076. Since the p-value is less than 0.05, we can reject the null hypothesis and conclude that there is a statistically significant difference in popularity between the genres. The t-value indicates the difference between the means of the two groups relative to the variance within the groups. A higher t-value indicates a greater difference between the means of the two groups.

# **Mann-Whitney U test**

**Is there a significant difference in the popularity of songs with different energy levels?**

 A Mann-Whitney U test is a non-parametric test that can be used to compare the means of two independent groups.

```python
import pandas as pd
from scipy.stats import mannwhitneyu
sub_df = copy_df.dropna(subset=["Popularity", "Energy"])
# Split the data into two groups based on energy level
high_energy = sub_df[sub_df['Energy'] > 0.712]['Popularity']
low_energy = sub_df[sub_df['Energy'] <= 0.712]['Popularity']

stat, p = mannwhitneyu(high_energy, low_energy)
print('Statistics=%.3f, p=%.3f' % (stat, p))
```

The result of Mann-Whitney U test shows that the test statistic is 12349741.000 and the p-value is 0.318. This means that there is no significant difference in the mean popularity of songs with high energy levels and songs with low energy levels.
